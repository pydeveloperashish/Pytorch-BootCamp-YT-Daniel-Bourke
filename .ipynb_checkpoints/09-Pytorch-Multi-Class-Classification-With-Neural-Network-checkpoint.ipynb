```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
NUM_CLASSES = 4    # no of classes
NUM_FEATURES = 2   # no of columns/features you want in dataset to predict a class.
RANDOM_SEED = 42
```

```python
# 1. Create multi-class data
X_blob, Y_blob = make_blobs(n_samples = 1000,
                            n_features = NUM_FEATURES,
                            centers = NUM_CLASSES,
                            cluster_std = 1.5,   # give the cluster a little shake up
                            random_state = RANDOM_SEED)
```

```python
X_blob[:10], Y_blob[:10]
```

```python
# 2. Convert data into Tensors
X_blob = torch.from_numpy(X_blob).type(torch.float)
Y_blob = torch.from_numpy(Y_blob).type(torch.LongTensor)
```

```python
print(X_blob.dtype)
print(Y_blob.dtype)
```

```python
# 3. Split train and test
X_train, X_test, Y_train, Y_test = train_test_split(X_blob, Y_blob, test_size = 0.2, random_state = RANDOM_SEED)
```

```python
len(X_train), len(Y_train), len(X_test), len(Y_test)
```

```python
# 4. Plot data (Visualize)
plt.figure(figsize = (10, 7))
plt.scatter(X_blob[:, 0], X_blob[:, 1], c = Y_blob, cmap = plt.cm.RdYlBu)
plt.show()
```

## 5. Building a Multi-Class Classification model in Pytorch

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
class BlobModel(nn.Module):
    def __init__(self, input_features, num_classes, hidden_units = 8):
        super().__init__()
        self.linear_layer_stack = nn.Sequential(
            nn.Linear(in_features = input_features, out_features = hidden_units),
            nn.ReLU(),
            nn.Linear(in_features = hidden_units, out_features = hidden_units),
            nn.ReLU(),
            nn.Linear(in_features = hidden_units, out_features = num_classes)
        )
        
    def forward(self, x):
        return self.linear_layer_stack(x)
    

model_0 = BlobModel(input_features = NUM_FEATURES, num_classes = NUM_CLASSES, hidden_units = 8)
print(model_0)
```

```python
X_train.shape, Y_train.shape
```

```python
model_0(X_train).shape
```

```python
# 6. Setup Loss and Optimizer

loss_fn = nn.CrossEntropyLoss()

optim = optim.SGD(model_0.parameters(), lr = 0.01)
```

```python
# Calculate accuracy - writing custom accuracy function
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item()
    acc = (correct / len(y_pred)) * 100
    return acc
```

In order to evaluate and train and test our model, we need to convert our model's outputs(logits) to prediction probabilities and then to prediction labels.

Logits (raw output of the model) -> Pred probs (use `torch.softmax`) -> Pred labels (take the argmax of the prediction probabilities)

```python
# 7. Build the Training and Testing Loop.

EPOCHS = 1000

X_train, Y_train = X_train.to(device), Y_train.to(device)
X_test, Y_test = X_test.to(device), Y_test.to(device)


for epoch in range(EPOCHS):
    
    model_0.train()   # Training mode ON
    
    # 1. Forward pass
    Y_logits = model_0(X_train)
    Y_pred = torch.softmax(Y_logits, dim = 1).argmax(dim = 1)  
    
    # 2. Calculate Loss
    train_loss = loss_fn(Y_logits, Y_train)
    
    train_acc = accuracy_fn(y_true = Y_train, y_pred = Y_pred)
    
    # 3. Reset the optimizer
    optim.zero_grad()
    
    # 4. Backpropogation
    train_loss.backward()
    
    # 5. Optimizer step (Gradient Descent)
    optim.step()
    
    
    ### Testing
    
    model_0.eval()     # Testing mode ON
    
    with torch.inference_mode():
        test_logits = model_0(X_test)
        test_preds = torch.softmax(test_logits, dim = 1).argmax(dim = 1)
        
        test_loss = loss_fn(test_logits, Y_test)
        
        test_acc = accuracy_fn(y_true = Y_test, y_pred = test_preds)
        
    # Print put what's happening
    if epoch % 100 == 0:
        print(f"Epoch: {epoch} | Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}% | Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%")
        
    
    
```

```python
# Check first 10 Logits
Y_logits[:10]
```

```python
# Go from Logits -> Predictions Probabilities
Y_pred_probs = torch.softmax(Y_logits, dim = 1)
Y_pred_probs[:10]
```

```python
# Go from Predictions Probabilities to Prediction Labels
Y_pred_labels = torch.argmax(Y_pred_probs, dim = 1)
Y_pred_labels[:10]
```
